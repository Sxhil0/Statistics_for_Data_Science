```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
file_path = r'C:\Users\Sahil Ayare\OneDrive\Desktop\Laliga.csv'
print(os.path.exists(file_path))
```

```python
data=pd.read_csv(r'C:\Users\Sahil Ayare\OneDrive\Desktop\Laliga.csv',index_col=0,header=1)
data
```

```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
data.replace("-",0,inplace=True)
data
```

```python
#Print all the teams which have started playing between 1930-1980.
data['Debut'] = data['Debut'].apply(lambda x: int(x.split('-')[0]))
data[(data['Debut'] >= 1930) & (data['Debut'] <= 1980)]
```

```python
#Print the list of teams which came Top 5 in terms of points
data['Points'] = data['Points'].astype(int)
data.sort_values(by="Points",ascending=False)
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference.
numeric_columns = ['GoalsFor', 'GoalsAgainst', 'Points', 'Seasons', 'GamesWon', 'GamesPlayed']
for column in numeric_columns:
    data[column] = pd.to_numeric(data[column])
def Goal_diff_count(data):
    data['GoalDifference'] = data['GoalsFor'] - data['GoalsAgainst']
    max_goal_diff_team = data.loc[data['GoalDifference'].idxmax()]
    min_goal_diff_team = data.loc[data['GoalDifference'].idxmin()]
    return data[['Team', 'GoalDifference']], max_goal_diff_team, min_goal_diff_team

goal_diff_df, max_goal_diff_team, min_goal_diff_team = Goal_diff_count(data)
print(goal_diff_df)
print("Team with max goal difference:", max_goal_diff_team['Team'])
print("Team with min goal difference:", min_goal_diff_team['Team'])
```

```python
#Create a new column with name “Winning Percent” and append it to the data set.
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which has the highest Winning percentage

# Calculate Winning Percent
data['WinningPercent'] = (data['GamesWon'] / data['GamesPlayed']) * 100
data['WinningPercent'].fillna(0, inplace=True)  # Replace any NaN with 0

# Print top 5 teams with the highest Winning percentage
top_5_winning_percent_teams = data.nlargest(5, 'WinningPercent')
print(top_5_winning_percent_teams[['Team', 'WinningPercent']])
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions 
best_position_points = data.groupby('BestPosition')['Points'].sum().reset_index()
print(best_position_points)
```
